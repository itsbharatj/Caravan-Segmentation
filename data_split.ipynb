#### Making the training and the validation split in the dataset. 
The second argument in the function `valid_size` denotes the percent of train dataset you want to convert into validation. By default the parameter is set to `20`% of the overall dataset

```python
import os 
import shutil
from PIL import Image 
import random
```

```python
'''
What needs to be done: 
- List all the files in image folder 
- Select 20% files from the list
- make a new folder 
- Move the respective files into the folder

- do the same for the masked images as well
'''
```

```python
len(os.listdir("/Users/bharatjain/Desktop/Deep Learning/UNet/dataset/train"))
```

```python
def data_split(data_dir="dataset",valid_size=20): 
    cwd = os.getcwd()
    images_folder = f'{data_dir}/train'
    print(images_folder)
    mask_folder = f'{data_dir}/train_masks'
    images = os.listdir(images_folder)
    n = int(len(images)*(valid_size/100))
    print(n)
    sample = random.sample(images,n)
    print(sample[1])
    if (not os.path.exists(f"{data_dir}/valid")):
        os.mkdir(f"{data_dir}/valid")
        os.mkdir(f"{data_dir}/valid_masks")
    for i in range(len(sample)):
        img_name = sample[i]
        # shutil.move(f"")
        img_path = os.path.join(images_folder,img_name)
        mask_path = os.path.join(mask_folder,img_name).replace(".jpg","_mask.gif")
        valid_img_path = f'{data_dir}/valid'
        valid_masks_path = f'{data_dir}/valid_masks'
        print(f'Image Path: {img_path}, Mask Path: {mask_path},Masks Path: {valid_masks_path}')
        shutil.move(img_path,valid_img_path)
        shutil.move(mask_path,valid_masks_path)
    print(f"Moved {len(sample)} images into the validation set")
```
